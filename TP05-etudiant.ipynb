> **(Polynôme interpolateur de Lagrange)**<br>
> Etant donnés $N$ un entier strictement positif, $x_0,\ldots,x_N$ des réels deux à deux distincts et $y_0,\ldots,y_N$ des réels, il existe un unique polynôme $P$ qui vérifie
> $$ P\in\mathbb{R}_{N}[X], \qquad P(x_i)=y_i, \quad 0\leq i\leq N.$$

---
### I : Construction en utilisant la base de Newton :  developpement  de Newton
---

**Rappel cours:**

Le polynôme interpolateur de Lagrange aux points $(x_i, y_i), i =0,\ldots,N$ est donné selon la formule de Newton par : 
$$
P = \sum_{i = 0}^{N} a_i \, w_i
$$
où les $w_i$ sont les polynômes de Newton définis par  
$$
\left\{ 
\begin{align}
w_0 &= 1 \\
w_1 &= X - x_0 \\
w_i &= (X - x_0)\ldots (X - x_{i-1}), \quad i = 2, \ldots N\quad \\
\end{align}
\right.
$$
et les coefficients $a_i$ sont obtenus par différences divisées: 
$
\quad \displaystyle a_i = d_{i,i}, \quad i = 0,\ldots,N
$

où les $(d_{i,j})$ définissent les entrées de la table des différences divisées 
et sont donnés par :
$$
\left\{ 
\begin{align}
d_{i,0} &= y_i  & \quad  0 \leq i \leq N& \qquad\\
d_{i, j} &= \frac{d_{i,j-1} - d_{i-1,j-1} }{x_i - x_{i-j}} &\quad  1\leq j \leq N, \quad j\leq i \leq N \quad\\
\end{align}
\right.
$$

Lorsque $y_i = f(x_i)$ (i.e lorsqu'on interpole une fonction $f$), on désigne $d_{i,j}$ par $f[x_{i-j},x_{i-j+1},\ldots,x_i]$. 

C'est-à-dire 
$
\displaystyle P = f[x_0] + f[x_0,x_1] (X - x_0) + f[x_0,x_1,x_2] (X - x_0)(X -x_1) + \ldots =  \sum_{i= 0}^N  f[x_{0},\ldots,x_i] \prod_{j = 0}^{i-1} (X - x_j)
$

On pourra remarquer (**voir cours**) que seuls les $d_{i,i}$ (éléments diagonaux de cette matrice) 
étant nécessaires on peut organiser le calcul des différences divisées de sorte à ne manipuler qu'un vecteur et non une matrice.
